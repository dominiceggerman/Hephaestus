```python
import pandas as pd
```

```python
pipes = pd.read_excel("Sample Pipe Rates.xlsx", sheet_name="Pipes")
caps = pd.read_excel("Sample Pipe Rates.xlsx", sheet_name="Capacities")
ie = pd.read_excel("Sample Pipe Rates.xlsx", sheet_name="IE Matrix")
pipe_names = ["TGP", "TRANSCO", "Stagecoach", "Dominion", "TCO", "TETCO", "Empire"]
```

```python
ie
```

```python
def step(ies):
    return [ind for ind, i in enumerate(ies) if i == 1]
```

```python
# Dom to TGP
# Dom -> Stage -> TGP
destination = False
pipe = "TGP"
dest = "Empire"

while destination == False:
    p = pipe_names.index(pipe)
    paths = step(ie.iloc[p].values)
    print("Current: {0}, Paths: {1}".format(pipe, list(pipe_names[x] for x in paths)))
    pipe = pipe_names[paths[0]]
    if pipe == dest:
        destination = True
```

```python
to_sell = 80000
df = pipes.set_index("Pipeline").join(caps.set_index("Pipeline"))
df
```

```python
# Calculate optimization
df["Deliverable"] = df["Lease Capacity"] - (df["Lease Capacity"] * df["Fuel"])
df["Cost"] = df["Lease Capacity"] * df["Rate"]
df["Sale"] = df["Deliverable"] * df["Price"]
df["Total Profit"] = df["Sale"] - df["Cost"]
df["Profit per Unit"] = df["Total Profit"] / df["Lease Capacity"]
df
```

```python
# Find sale volumes
sold = []
for vol in df["Lease Capacity"].values:
    # Break if no volume remaining
    if to_sell - vol <= 0:
        sold.append(to_sell)
        break
    # Append and update
    sold.append(vol)
    to_sell = to_sell - vol
# Append sold volumes to df
sold = pd.DataFrame(sold, columns=["Vol Sold"])
df = df.reset_index().join(sold)
df
```

```python
# Calculate total profit
df.dropna(axis=0, inplace=True)
print(df["Total Profit"].sum())
```
